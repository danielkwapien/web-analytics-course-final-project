# Prepare the data

## 1. Import and clean the data

First we will import and clean the data extracted from the TicketMaster API.

```python
import pandas as pd
```

```python
events_df = pd.read_csv('./data/events_usa.csv')
events_df
```

```python
events_df = events_df[events_df['status'] == 'onsale']
events_df = events_df.dropna()
events_df
```

At the beggining we had too much data because we collected all the events from the API. We filtered the data to only include events that are on sale and have a price. We also removed the events that have any missing values. With that, we are left with 43,502 events.

## 2. Prepare the datasets

Now we will prepare the datasets we need to be used in the streamlit app

```python
events_df['average_price'] = events_df[['min_price', 'max_price']].mean(axis=1)
```

```python
events_df.to_csv('./data/events.csv', index=False)
```

```python
prices_by_city = events_df.groupby(['venue_city', 'segment_name']).agg({'min_price':'mean', 'max_price':'mean', 'average_price':'mean', 'venue_latitude':'mean', 'venue_longitude':'mean'}).reset_index()
prices_by_city.to_csv('./data/prices_by_city.csv', index=False)
```

```python
state_name_to_abbreviation = {
    'Alabama': 'AL', 'Alaska': 'AK', 'Arizona': 'AZ', 'Arkansas': 'AR',
    'California': 'CA', 'Colorado': 'CO', 'Connecticut': 'CT', 'Delaware': 'DE',
    'Florida': 'FL', 'Georgia': 'GA', 'Hawaii': 'HI', 'Idaho': 'ID',
    'Illinois': 'IL', 'Indiana': 'IN', 'Iowa': 'IA', 'Kansas': 'KS',
    'Kentucky': 'KY', 'Louisiana': 'LA', 'Maine': 'ME', 'Maryland': 'MD',
    'Massachusetts': 'MA', 'Michigan': 'MI', 'Minnesota': 'MN', 'Mississippi': 'MS',
    'Missouri': 'MO', 'Montana': 'MT', 'Nebraska': 'NE', 'Nevada': 'NV',
    'New Hampshire': 'NH', 'New Jersey': 'NJ', 'New Mexico': 'NM', 'New York': 'NY',
    'North Carolina': 'NC', 'North Dakota': 'ND', 'Ohio': 'OH', 'Oklahoma': 'OK',
    'Oregon': 'OR', 'Pennsylvania': 'PA', 'Rhode Island': 'RI', 'South Carolina': 'SC',
    'South Dakota': 'SD', 'Tennessee': 'TN', 'Texas': 'TX', 'Utah': 'UT',
    'Vermont': 'VT', 'Virginia': 'VA', 'Washington': 'WA', 'West Virginia': 'WV',
    'Wisconsin': 'WI', 'Wyoming': 'WY'
}
```

```python
events_df["venue_state_code"] = events_df["venue_state"].map(state_name_to_abbreviation)
```

```python
prices_by_state = events_df.groupby(['venue_state_code', 'segment_name', 'genre_name', 'sub_genre_name']).agg({'min_price':'mean', 'max_price':'mean', 'average_price':'mean'}).reset_index()
prices_by_state.to_csv('./data/prices_by_state.csv', index=False)
```

```python
pre_events_df = events_df[events_df['genre_name'] != 'Undefined']
prices_by_genre = pre_events_df.groupby(['segment_name', 'genre_name', 'sub_genre_name']).agg(average_price=("average_price", "mean"),count=("min_price", "count")).reset_index()
pre_events_df = events_df[events_df['segment_name']=='Music']
prices_by_genre.to_csv('./data/prices_by_genre.csv', index=False)
```
