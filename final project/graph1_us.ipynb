```python
# old funciton
import networkx as nx

def create_graph(artist_genre_data):
    """
    create tehe Artist-Genre graph
    two types of nodes: artist and genre
    edges between artist and genre
    """
    # Create a new graph
    G = nx.Graph()

    # Add nodes and edges to the graph
    for _, row in artist_genre_data.iterrows():
        artist = row['artist_name']
        genre = row['Genre']

        # Add nodes for artist and genre
        G.add_node(artist, type='artist')
        G.add_node(genre, type='genre')

        # Add an edge between the artist and the genre
        G.add_edge(artist, genre)

    return G
```

```python
# new funtion 

def create_graph(genre_category, artist_genre_data):
    """
    create tehe Artist-Genre graph
    two types of nodes: artist and genre
    edges between artist and genre
    """

    data = artist_genre_data[artist_genre_data['genre_category'] == genre_category]

    # Create a new graph
    G = nx.Graph()

    # Add nodes and edges to the graph
    for _, row in data.iterrows():
        artist = row['attraction_name']
        genre = row['genre_name']
        subgenre = row['sub_genre_name']

        # Add nodes for artist and genre
        G.add_node(artist, type='artist')
        G.add_node(genre, type='genre')

        # add subgenre info to the genre node
        G.nodes[genre]['sub_genre'] = subgenre

        # Add an edge between the artist and the genre
        G.add_edge(artist, genre)

    return G
```

```python
def interactive_plot(G, genre_category):
    # Get positions for nodes
    pos = nx.spring_layout(G)

    # Prepare node traces for Plotly
    node_x = []
    node_y = []
    node_labels = []
    node_hover = []
    node_colors = []

    for node in G.nodes():
        x, y = pos[node]
        node_x.append(x)
        node_y.append(y)

        # Check node type and assign hover text and color accordingly
        if G.nodes[node].get('type') == 'artist':
            # Get the genres and subgenres for the artist
            genres = [neighbor for neighbor in G.neighbors(node) if G.nodes[neighbor].get('type') == 'genre']
            subgenres = [G.nodes[genre].get('sub_genre') for genre in genres]
            hover_text = f"Artist: {node}<br>Genres: {', '.join(genres)}"
            if any(subgenre != genre for genre, subgenre in zip(genres, subgenres)):
                hover_text += f"<br>Subgenres: {', '.join(subgenre for genre, subgenre in zip(genres, subgenres) if subgenre != genre)}"
            node_hover.append(hover_text)
            node_colors.append('skyblue')
        elif G.nodes[node].get('type') == 'genre':
            node_hover.append(f"Genre: {node}")
            node_colors.append('orange')
        else:
            node_hover.append(f"Node: {node}")
            node_colors.append('gray')  # Default color if type is undefined

        # Use the node name as the label
        node_labels.append(node)

    node_trace = go.Scatter(
        x=node_x,
        y=node_y,
        mode='markers',
        marker=dict(
            size=10,
            color=node_colors,
            line_width=1
        ),
        text=node_labels,
        hovertext=node_hover,
        hoverinfo="text"
    )

    # Prepare edge traces for Plotly
    edge_x = []
    edge_y = []

    for edge in G.edges():
        x0, y0 = pos[edge[0]]
        x1, y1 = pos[edge[1]]
        edge_x.append(x0)
        edge_x.append(x1)
        edge_x.append(None)
        edge_y.append(y0)
        edge_y.append(y1)
        edge_y.append(None)

    edge_trace = go.Scatter(
        x=edge_x,
        y=edge_y,
        line=dict(width=0.5, color="#888"),
        hoverinfo="none",
        mode="lines"
    )

    # Create the Plotly figure
    fig = go.Figure(data=[edge_trace, node_trace],
                    layout=go.Layout(
                        title=f"Performer-Genre Interactive Graph ({genre_category})",
                        titlefont_size=16,
                        showlegend=False,
                        hovermode="closest",
                        margin=dict(b=0, l=0, r=0, t=40),
                        xaxis=dict(showgrid=False, zeroline=False),
                        yaxis=dict(showgrid=False, zeroline=False)
                    ))

    return fig

```

```python
def calculate_genre_centrality(G):
    """
    Calculates degree centrality for genre nodes and returns
    all genres sorted by centrality in descending order, excluding 'Other'.
    """
    # Compute degree centrality
    centrality = nx.degree_centrality(G)

    # Filter for genre nodes and exclude 'Other'
    genre_centrality = {
        node: centrality[node]
        for node, attr in G.nodes(data=True)
        if attr.get("type") == "genre" and node != "Other" # exclude other from the genres
    }

    # Sort genres by centrality in descending order
    sorted_genres = sorted(genre_centrality.items(), key=lambda x: x[1], reverse=True)

    # Return sorted list of genres
    return sorted_genres

```

```python
import networkx as nx

def get_artist_betweenness_centrality(graph):
    """
    Calculate the betweenness centrality of artists in a bipartite artist-genre graph.
    
    Parameters:
        graph (nx.Graph): The artist-genre bipartite graph.
    
    Returns:
        dict: A dictionary with artists as keys and their betweenness centrality as values.
    """
    # Filter the nodes to only include artists
    artists = [node for node, attrs in graph.nodes(data=True) if attrs.get('type') == 'artist']
    
    # Compute betweenness centrality for all nodes in the graph
    centrality = nx.betweenness_centrality(graph)
    
    # Filter the centrality dictionary to include only artists
    artist_centrality = {node: centrality[node] for node in artists}
    
    return artist_centrality
```

```python
def get_genres_connected_to_high_centrality_artists(graph, centrality_threshold=0.1):
    """
    Identify genres connected to artists with high betweenness centrality.
    
    Parameters:
        graph (nx.Graph): The artist-genre bipartite graph.
        centrality_threshold (float): Minimum betweenness centrality for an artist to be considered highly central.
        
    Returns:
        dict: A dictionary where keys are genres, and values are lists of artists connected to those genres.
    """
    # Get artist betweenness centrality using the previously defined function
    artist_centrality = get_artist_betweenness_centrality(graph)
    
    # Filter artists with high betweenness centrality
    high_centrality_artists = [
        artist for artist, centrality in artist_centrality.items() if centrality > centrality_threshold
    ]
    
    # Create a dictionary to store genres and their associated high-centrality artists
    genre_to_artists = {}
    
    for artist in high_centrality_artists:
        # Get all genres connected to this artist
        genres = [neighbor for neighbor in graph.neighbors(artist) if graph.nodes[neighbor].get('type') == 'genre']
        
        for genre in genres:
            if genre not in genre_to_artists:
                genre_to_artists[genre] = []
            genre_to_artists[genre].append(artist)
    
    return genre_to_artists
```

```python
def get_common_genre_pairs_and_artists(graph):
    """
    Find pairs of genres that share common artists and the artists they share.
    
    Parameters:
        graph (nx.Graph): The artist-genre bipartite graph.
        
    Returns:
        dict: A dictionary where keys are tuples of genre pairs (genre1, genre2),
              and values are lists of artists shared by those genres.
    """
    # Filter the nodes to only include genres
    genres = [node for node, attrs in graph.nodes(data=True) if attrs.get('type') == 'genre']
    
    # Create a dictionary to store genre pairs and their common artists
    genre_pairs = {}
    
    # Iterate through all combinations of genres
    for i, genre1 in enumerate(genres):
        for genre2 in genres[i + 1:]:
            # Find common artists between the two genres
            artists_genre1 = set(neighbor for neighbor in graph.neighbors(genre1) if graph.nodes[neighbor].get('type') == 'artist')
            artists_genre2 = set(neighbor for neighbor in graph.neighbors(genre2) if graph.nodes[neighbor].get('type') == 'artist')
            
            common_artists = artists_genre1.intersection(artists_genre2)
            
            if common_artists:
                genre_pairs[(genre1, genre2)] = list(common_artists)
    
    return genre_pairs
```
